# Detection Model Training

Train the detection model offline, then export the weights for runtime inference.

The exported artifact is saved to `artifacts/detection_model/model.pt`.

```python
from model import OWASPSystem, Config

config = Config(
    device='cpu',
    enable_online_learning=False,
    model_artifact_dir='artifacts/detection_model',
)
system = OWASPSystem(config)

# Update these paths to your local training datasets.
paths = {
    'csic': '/home/hunter/Final_P/cyberguard/test11/csic_database.csv',
    'cic': '/not/available.csv',
    'payloads_folder': '/not/available_folder',

}

texts, labels = system.load_data(paths)
print(f'Total samples: {len(texts)}')
system.train_initial(texts, labels)
print('Saved model to:', system._weights_path())
print('Saved metadata to:', system._metadata_path())
```

```python
# Optional quick inference check after training.
reloaded = OWASPSystem(Config(device='cpu', enable_online_learning=False))
result = reloaded.predict_request('http://example.com', '/item?id=1 UNION SELECT')
result
```
